```python
data_1 = pd.read_csv(r'C:\Users\Матвей\Downloads\books.csv\ratings.csv', sep=";", 
                    error_bad_lines=False, encoding="latin-1")
#загружаем данные с рейтингом от пользователей
```

```python
rating = pd.DataFrame(data=data_1)
#создаём новый датафрейм
```

```python
rating_df = rating.rename(columns={'User-ID':'user_id','Book-Rating':'book_rating'})
```

```python
rating_df
```

```python
rating_df.book_rating.unique()
#получаем список всех уникальных значений
#так мы можем увидеть, если какие-то значений отстутствуют, в принципе
#также можно понять нет ли сторонних значенй, как в случае с предыдущим датафреймом
```

```python
rating_df.mode(numeric_only=True)
#numeric_only считает моду только в числовых данных
#так мы видим, что чаще всего встречается пользователь под user-id 11676
#а самая часто встречающаяся оценка 0
```

```python
len(rating_df[rating_df['book_rating']==0])
# смотрим, сколько раз в датафрейме встречается значение 0 в столбце рейтинга
```

```python
rating_only_df = rating_df[['ISBN','book_rating']]
# формируем новый датасет. здесь не прибегаем к удалению столбцов из старого
# потому что нам еще потребуется та информация
```

```python
rating_only_df.groupby('ISBN').mean()
```

```python
full_books_df = pd.merge(books_df, rating_only_df, on="ISBN", how="inner")
# создаём новый датафрейм, где соединяем два предыдущих
```

```python
full_books_df =  np.round(full_books_df, decimals=1)
# количество строк сильно увеличилось, так как в нолвый датафрейм был получен путем, мерджа по 
# ISBN. Для того, чтобы построить график и не потерять значения, сгрупппируем даныне по авторам
# и получим их суммарное кол-во книг и рейтинг
```

```python
full_books_df
```

```python
result_df = full_books_df[full_books_df.sort_values(by='book_rating', ascending=False)['book_rating'] > 0]
# убираем значения, которые имеют рейтинг 0. т.к они не влияют на общую сумму.
# и, скорее всего, значение 0 присваивается также не оценненным книгам. 
```

```python
result_df
```

```python
grouped_result_df = result_df.groupby(by='book_author',as_index = False)
```

```python
grouped_result_df = grouped_result_df.count().sort_values('book_title', ascending=False)
# получаем кол-во книг, названий и прочего
```
